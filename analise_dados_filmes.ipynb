```python
import pandas as pd

table_filmes = pd.read_csv("movie_dataset.csv")
display(table_filmes)
```

### Tratando e Analisando Valores Nulos

```python
table_filmes['genres'] = table_filmes['genres'].fillna('without genres')

table_filmes['homepage'] = table_filmes['homepage'].fillna('no home page')

table_filmes['keywords'] = table_filmes['keywords'].fillna('no keywords')

table_filmes['tagline'] = table_filmes['tagline'].fillna('no tagline')

display(table_filmes)
```

### Tratando e Analisando a quantidade de filmes por gen√™ros

```python
qnt_filmes_genero = table_filmes['genres'].value_counts()
display(qnt_filmes_genero)
```

### Tratando e Analisando o or√ßamento por categoria

```python
orcamento_genero = table_filmes[['budget', 'genres']].groupby(['genres']).sum()
display(orcamento_genero)
```

### An√°lisando e Tratando A soma dos Votos e M√©dia por Gen√™ro/Filme

```python
votos_total_media = table_filmes[['genres', 'title', 'vote_average', 'vote_count']].groupby(['genres', 'title']).sum()
display(votos_total_media)
```

### Analisando e Tratando Filmes com a Nota > 5

```python
filmes_nota_maior_5 = table_filmes.loc[table_filmes['vote_average'] > 5.0, ['genres', 'title', 'vote_average']]
display(filmes_nota_maior_5)
```

### Analisando e Filmes com a Nota < 5

```python
filmes_nota_menor_5 = table_filmes.loc[table_filmes['vote_average'] < 5.0, ['genres', 'title', 'vote_average']]
display(filmes_nota_menor_5)
```
